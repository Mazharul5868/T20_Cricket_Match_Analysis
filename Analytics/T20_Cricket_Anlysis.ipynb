```python
import pandas as pd 
import json 
```

```python
with open ('t20_json_files/t20_wc_match_results.json') as f: 
    data = json.load(f)
data
```

```python
pd.DataFrame(data[0]['matchSummary'])
```

```python
df_match = pd.DataFrame(data[0]['matchSummary']) 
df_match.head()
```

```python
df_match.shape
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows(): 
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

```python
df_match.shape
```

```python
df_match.rename(columns={'scorecard' : 'match_id'}, inplace=True)
```

```python
df_match.head()
```

```python
df_match.to_csv('t20_csv_files/t20_wc_match_results.csv', index=False)
```

**Batting Summary**

```python
with open ('t20_json_files/t20_wc_batting_summary.json') as f: 
    data = json.load(f)

    all_records = []
    
    for rec in data:
        all_records.extend(rec['battingSummary'])
all_records
```

```python
df_batting = pd.DataFrame(all_records)
df_batting
```

```python
df_batting['out/not_out'] = df_batting['dismissal'].apply(lambda x: 'Out' if len(x) > 0 else 'not_out')
df_batting
```

```python
df_batting.drop(columns=['dismissal'], inplace=True) 
df_batting
```

```python
df_batting.head(50)
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x : x.replace('\xa0', ''))
df_batting.head(20)
```

```python
df_batting.head(30)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/t20_wc_batting_summary.csv', index=False)
```
